```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv.imread('/content/drive/MyDrive/PCVK/female.tiff', 0)  # 0 untuk membaca dalam skala abu-abu (grayscale)

# Buat Structuring Element Dilasi (SED) berbentuk square 5x5
sed = np.ones((5, 5), np.uint8)

# Fungsi dilasi
def dilate(image, sed):
    m, n = image.shape
    y, x = sed.shape
    y = y // 2
    x = x // 2
    dilated_image = np.zeros((m, n), dtype=np.uint8)

    for i in range(y, m - y):
        for j in range(x, n - x):
            dilated_image[i, j] = np.max(image[i - y:i + y + 1, j - x:j + x + 1] * sed)

    return dilated_image

# Terapkan dilasi ke gambar
dilated_image = dilate(image, sed)

cv2_imshow(dilated_image)
```
