**2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!**

```python
red_channel = gambarPolosan[:, :, 2]
blue_channel = gambarPolosan[:, :, 0]


rb_image = np.zeros_like(gambarPolosan)
rb_image[:, :, 0] = red_channel
rb_image[:, :, 1] = blue_channel


plt.imshow(rb_image)
```

```python
green_channel = gambarPolosan[:, :, 1]
blue_channel = gambarPolosan[:, :, 0]


gb_image = np.zeros_like(gambarPolosan)
gb_image[:, :, 1] = green_channel
gb_image[:, :, 0] = blue_channel

plt.imshow(gb_image)
```
